```python
import random as rd

class Academia:
    def __init__(self):
        self.halteres = [i for i in range(10, 36) if i % 2 == 0]
        self.portaHalteres = {}
        self.reiniciar_o_dia()

    def reiniciar_o_dia(self):
        self.portaHalteres = {i: i for i in self.halteres}

    def listar_espaco(self):
        return [pos for pos, peso in self.portaHalteres.items() if peso == 0]

    def listar_halteres(self):
        return [peso for peso in self.portaHalteres.values() if peso != 0]

    def pegarHalteres(self, peso):
        for pos, valor in self.portaHalteres.items():
            if valor == peso:
                self.portaHalteres[pos] = 0
                return peso
        return 'Peso indisponivel'

    def devolver_halter(self, pos, peso):
        self.portaHalteres[pos] = peso

    def calcular_caos(self):
        num_caos = [i for i, j in self.portaHalteres.items() if i != j]
        return len(num_caos) / len(self.portaHalteres)


class Usuario:
    def __init__(self, tipo, academia):
        self.tipo = tipo  # 1 - Normal | 2 - Bagunceiro
        self.academia = academia
        self.peso = 0

    def iniciar_treino(self):
        lista_pesos = self.academia.listar_halteres()
        self.peso = rd.choice(lista_pesos)
        self.academia.pegarHalteres(self.peso)

    def finalizar_treino(self):
        espacos = self.academia.listar_espaco()

        # Tipo 1 → tenta devolver no lugar certo
        if self.tipo == 1:
            pos_correta = self.peso
            if pos_correta in espacos:
                self.academia.devolver_halter(pos_correta, self.peso)
            else:
                pos = rd.choice(espacos)
                self.academia.devolver_halter(pos, self.peso)

        # Tipo 2 → devolve em qualquer lugar
        elif self.tipo == 2:
            pos = rd.choice(espacos)
            self.academia.devolver_halter(pos, self.peso)

        self.peso = 0


```

```python
academia = Academia()
usuarios = [Usuario(1, academia) for i in range(10)]
usuarios += [Usuario(2, academia) for i in range(1)]
rd.shuffle(usuarios)

taxa_treino = {}
for i in range(7):
    rd.shuffle(usuarios)
    for user in usuarios:
        user.iniciar_treino()
    for user in usuarios:
        user.finalizar_treino()
        taxa_treino[i] = academia.calcular_caos()

for u, c in taxa_treino.items():
    print(f'Para o treino {u + 1}, a taxa de bagunça é de {c * 100:.1f}')


```
